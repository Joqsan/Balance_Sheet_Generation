```python
dataframes = {
    'AFP' : [],
    'FONASA' : [],
    'ISAPRE' : [],
    'MUTUAL' : [],
    'CCAF' : []
}
```

```python
for folder in folders:
    folder_path = '/Users/joqsan/0-Excel/Reading_From_Folders/' + folder + '/*.pdf'
    files = [filename for filename in glob.glob(folder_path)]
    
    for file in files:
        dfs = camelot.read_pdf(file, pages='1', flavor='stream', 
                               table_areas=params[folder]['area'], columns=params[folder]['cols'])
        df = params[folder]['dc_function'](dfs)
        dataframes[folder].append(df)
```

```python
for entity, df_list in dataframes.items():
    dataframes[entity] = pd.concat(df_list)
    dataframes[entity] = sort_index(dataframes[entity])
```
